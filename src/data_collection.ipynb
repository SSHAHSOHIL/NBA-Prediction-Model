```python
from nba_api.stats.endpoints import LeagueDashTeamStats
from nba_api.stats.endpoints import LeagueGameLog
import pandas as pd
```

```python
df_team = (
    LeagueDashTeamStats (
        season= '2023-24',
        season_type_all_star='Regular Season'
    )
    .get_data_frames()[0]
)
```

```python
df_games_1 = (
    LeagueGameLog(
        season = '2024-25',
        season_type_all_star='Regular Season'
    ).get_data_frames()[0]
)

df_games_2 = (
    LeagueGameLog(
        season = '2024-25',
        season_type_all_star='Playoffs'
    ).get_data_frames()[0]
)

df_games = pd.concat([df_games_1, df_games_2], ignore_index=True)

df_games
```

```python
df_games['WIN'] = (df_games['WL'] == 'W').astype(int)
print(df_games)
```

```python
features = ['TEAM_ID', 'W_PCT', 'PTS_RANK', 'AST_RANK', 'REB_RANK']
season_team_features = df_team[features]
print(season_team_features)
```

```python
df_games_with_season_stats = df_games.merge (
    season_team_features,
    on = 'TEAM_ID',
    how = 'left'
)
```

```python
df_games_with_season_stats
```

```python
from nba_api.stats.static import teams
team_name_to_id = {}
for t in teams.get_teams():
    team_name_to_id[t['abbreviation']] = t['id']
```

```python
# team, vs, opp = df_games_with_season_stats['MATCHUP'][0].split()

# opponent ID and home advantage

new_col = []
home_col = []
for line in df_games_with_season_stats['MATCHUP']:
    team, vs, opp = line.split()
    if vs != '@':
        home_col.append(1)
    else:
        home_col.append(0)
    new_col.append(team_name_to_id[opp])

df_games_with_season_stats['OPP_ID'] = new_col
df_games_with_season_stats['is_home'] = home_col


```

```python
def elo_win_prob(elo_a, elo_b):
    pt1 = 1 + pow(10, (elo_b - elo_a)/400)
    return 1 / pt1
```

```python
# elo rankings for the teams as the season goes on

team_elos = {}
for id in df_games_with_season_stats['TEAM_ID'].unique():
    team_elos[id] = 1500


K = 20 # learning rate, essentially how dramatically the ranking moves game by game
# whether the team loses or wins

pre_elo = []
pre_opp_elo = []
df_games_with_season_stats = df_games_with_season_stats.sort_values(['GAME_DATE','TEAM_ID']).reset_index(drop=True)

for _, r in df_games_with_season_stats.iterrows():
    team_id = r['TEAM_ID']
    opp_id = r['OPP_ID']

   
    e_t = team_elos[team_id]
    e_o = team_elos[opp_id]
    
    pre_elo.append(e_t)
    pre_opp_elo.append(e_o)

    expected_prob = elo_win_prob(e_t, e_o)
    s_a = r['WIN']


    team_elos[team_id] = e_t + K * (s_a - expected_prob)
    team_elos[opp_id] = e_o + K * ((1 - s_a) - (1 - expected_prob))

    # s_a is the actual value (whether the team won or lost)
    # expected_prob is the expected score from the formula based on the teams ELOS
    # a team will outperform its expectation if s_a - expected_prob is greater than 0
    #     as a result the team's ELO would go up, otherwise it will go down
    
df_games_with_season_stats['elo'] = pre_elo
df_games_with_season_stats['elo_opp'] = pre_opp_elo
df_games_with_season_stats['elo_diff'] = df_games_with_season_stats['elo'] - df_games_with_season_stats['elo_opp']


```

```python
# momentum
df_games_with_season_stats['momentum_5'] = (
    df_games_with_season_stats.groupby('TEAM_ID')['WIN']
        .rolling(window = 5, min_periods=1)
        .mean()
        .shift(1) # prevent leakage of knowing the Game N's result
        .reset_index(level=0, drop=True)
)

df_games_with_season_stats['momentum_5'] = df_games_with_season_stats['momentum_5'].fillna(0)
```

```python
# win streak
win_streak = []
map_to_streak = {}
keys = []
for _, r in df_games_with_season_stats.iterrows():
    keys.append(r['TEAM_ID'])

map_to_streak = dict.fromkeys(keys, 0)
print(map_to_streak)

for _, r in df_games_with_season_stats.iterrows():
    team_id = r['TEAM_ID']
    result = r['WIN']
    
    if result == 1:
        new_val = map_to_streak[team_id] + 1
        win_streak.append(new_val)
        map_to_streak[team_id] = new_val
    else:
        win_streak.append(0)
        map_to_streak[team_id] = 0

df_games_with_season_stats['win_streak'] = win_streak



```

```python
df_games_with_season_stats = df_games_with_season_stats.merge(
    season_team_features,
    left_on='OPP_ID',
    right_on='TEAM_ID',
    suffixes=('','_opp')
)
```

```python
def get_opponent_score(matchup, game_date, df):
    """
    Retrieves the opponent's score from the DataFrame.

    Args:
        matchup (str): The matchup string (e.g., 'TeamA vs TeamB').
        game_date (str): The game date.
        df (pd.DataFrame): The DataFrame containing game data.

    Returns:
        int: The opponent's score, or None if not found.
    """
    team_1, _, team_2 = matchup.split()
    rows_with_value = df[(df['TEAM_ABBREVIATION'] == team_2) & (df['GAME_DATE'] == game_date)]
    if not rows_with_value.empty:
        # .iloc[0] is safer than just accessing the first element.  Handles the case where the query returns multiple rows (it shouldn't, but better to be safe).
        return rows_with_value['PTS'].iloc[0]
    else:
        return None  # Handle cases where the opponent's score isn't found.  Important for data integrity.


# Apply the function to create the 'PTS_AWAY' column
df_games_with_season_stats['PTS_AWAY'] = df_games_with_season_stats.apply(
    lambda row: get_opponent_score(row['MATCHUP'], row['GAME_DATE'], df_games_with_season_stats), axis=1
)
```
